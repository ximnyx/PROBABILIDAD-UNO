### 1. PROBLEMA DEL CUMPLEA√ëOS 
Supongamos que n personas no relacionadas entre s√≠, son invitadas a una fiesta y que cada persona tiene la misma probabilidad de nacer cualquier d√≠a del a√±o (no a√±os bisiestos). 


    a) ¬øCu√°l es la probabilidad que dos o m√°s invitados cumplan a√±os el mismo d√≠a?

```python
import math

def prob_cumple_compartido(n):
    prob_distintos = 1
    for k in range(n):
        prob_distintos *= (365 - k) / 365
    return 1 - prob_distintos

n = 23
print("Probabilidad:", "{:.4f}".format(prob_cumple_compartido(n)))
```

    b) ¬øCu√°l es la m√≠nima cantidad para que la probabilidad sea de al menos 50%(es decir que al menos dos personas nazcan el mismo d√≠a)?

```python
def minimo_n_50():
    n = 1
    while prob_cumple_compartido(n) < 0.5:
        n += 1
    return n, prob_cumple_compartido(n)

n_min, prob = minimo_n_50()
print("n m√≠nimo:", n_min)
print("Probabilidad:", "{:.4f}".format(prob))
```

### 6. OTRO EJERCICIO DE BOLAS EN UNA URNA
Se van colocando al azar bolas, una a una, en cualquiera de N cajas hasta que alguna caja llegue a tener dos bolas.<br>

    a) ¬øcu√°l es la probabilidad de que el proceso termine en el paso n con 2‚â§n‚â§N+1?
*Soluci√≥n*: consideremos los siguientes eventos. Sea $n  œµ${$2,‚Ä¶,n$} <br>
A: el proceso termina en el paso n<br>
B<sub>k</sub>: la k-√©sima bola se coloca en una caja desocupada<br>
B<sub>n</sub>: la n-√©sima bola se coloca en una caja ocupada<br>







```python
from fractions import Fraction

def prob_termina_en_n(N, n):
    if n < 2 or n > N + 1:
        return None

    p = Fraction(1, 1)

    for k in range(1, n - 1):
        p *= Fraction(N - k, N)

    p *= Fraction(n - 1, N)
    return p


# Ejemplo
N = 10
n = 5

p = prob_termina_en_n(N, n)

print(f"Probabilidad de terminar en el paso {n}:", p, "=", "{:.3f}".format(float(p)))
```

### 7. PROBLEMAS DE LAS CARTAS [·¥á·¥†·¥á…¥·¥õ·¥èÍú± …™…¥·¥Ö·¥á·¥ò·¥á…¥·¥Ö…™·¥á…¥·¥õ·¥áÍú±]
Supongamos que una carta es elegida aleatoriamente de mazo de 52 cartas. Determine si los siguientes eventos son independientes.<br>

Definamos:<br>
A: la carta obtenida es un As<br>
B: la carta obtenida es un tr√©bol<br>

Calcular las probabilidades de:<br>
$‚Ñô(A)= \frac{4}{52} = \frac{1}{13}$ <br>
$‚Ñô(B)= \frac{13}{52}= \frac{1}{4}  $ <br>
$‚Ñô(A ‚à© B)= \frac{1}{52}$ <br>

$‚Ñô(A)‚Ñô(B)= \frac{1}{13} \cdot \frac{1}{4}= \frac{1}{52}$ <br>
por lo tanto, $‚Ñô(A ‚à© B)= ‚Ñô(A) ‚Ñô(B)$, y los eventos A y B son independientes.<br>


```python
from fractions import Fraction


P_A = Fraction(4, 52)      
P_B = Fraction(13, 52)    
P_A_inter_B = Fraction(1, 52)  

print("P(A):", P_A)
print("P(B):", P_B)
print("P(A ‚à© B):", P_A_inter_B)
print("P(A)P(B):", P_A * P_B)

print("¬øSon independientes?", P_A_inter_B == P_A * P_B)
```

### 9. PROBLEMA DEL EXAMEN DE OPCI√ìN M√öLTIPLE [·¥ò Ä·¥è ô·¥Ä ô…™ ü…™·¥Ö·¥Ä·¥Ö  ô·¥Ä è·¥áÍú±…™·¥Ä…¥·¥Ä]
Supongamos que un examen de opci√≥n m√∫ltiple, cada pregunta tiene 5 opciones de las cuales el estudiante elige una como la respuesta correcta. <br>
Un estudiante conoce la respuesta con probabilidad de 0.7 y elige aleatoriamente una de las 5 alternativas como su respuesta correcta en caso <br>
de no conocer la respuesta correcta.<br>

Supongamos que una pregunta en particular fue respondida correctamente.<br>
1. ¬øCu√°l es la probabilidad de que el estudiante conociera realmente la respuesta correcta?<br>
Definamos los siguientes dos eventos:<br>
$A$: EL estudiante conoc√≠a realmente la respuesta.<br>
$B$: EL estudiante respondi√≥ correctamente la pregunta.<br>

Usando Bayes: <br>
$P(A‚îÇB)= \frac{P(B‚îÇA)P(A)}{P(B‚îÇA)P(A)+P(B‚îÇA^C )P(A^C )}$ <br>
$P(A)=0.7$<br>
$P(B‚îÇA)=1$<br>
$P(A^C )=0.3$<br>
$P(B‚îÇA^C )=1/5$<br>

Sustituyendo dados:<br>
$=\frac{0.7}{0.7+ \frac{1}{5}(0.3)} =0.9211$<br>


```python
P_A = 0.7
P_B_dado_A = 1
P_Ac = 0.3
P_B_dado_Ac = 1/5

P_A_dado_B = (P_B_dado_A * P_A) / (P_B_dado_A * P_A + P_B_dado_Ac * P_Ac)

print(f"P(A | B) = {P_A_dado_B:.4f}")
```

### 10. PROBLEMA DE UN ESTUDIO DE SANGRE
Un estudio de sangre para una enfermedad espec√≠fica, nos da un resultado positivo un 90% de las veces entre pacientes que tienen la enfermedad, pero <br>
tambi√©n arroja un resultado positivo un 25% de las veces entre persnas ue no tienen la enfermedad. Se sabe adem√°s que 30% de la poblaci√≥n de estudio est√° enferma.<br>

    1.¬øCu√°l es la probabilidad de que una persona con un resultado positivo de la prueba realmente tenga la enfermedad?

Definamos los siguientes eventos:<br>
$A$: La persona est√° enferma.<br>
$B$: La persona obtuvo un resultado positivo en la prueba de sangre.<br>

$P(A‚îÇB)= \frac{P(B‚îÇA)P(A)}{P(B‚îÇA)P(A)+P(B|A^C)P(A^C)}  = \frac{0.9(0.3)}{0.9(0.3)+0.25(0.7)}=0.6067$ <br>
 
$P(B‚îÇA)=0.9, P(A)=0.3, P(A^C )=0.7, P(B‚îÇA^C )=0.25$<br>



```python
P_A = 0.3
P_Ac = 0.7
P_B_dado_A = 0.9
P_B_dado_Ac = 0.25

P_A_dado_B = (P_B_dado_A * P_A) / (P_B_dado_A * P_A + P_B_dado_Ac * P_Ac)

print(f"P(A|B) = {P_A_dado_B:.4f}")
```

### 11. PROBLEMA DE URNAS
Supongamos que se tienen dos urnas: 1 y 2. Tal que la urna 1 contiene una bola negra y una bola blanca,mientras que la urna 2 contiene dos bolas negras y tres bolas blancas.<br>
Una de las dos urnas es elegida aleatoriamente y una bola es extra√≠da de esta. Si una bola negra es extra√≠da , ¬øcu√°l es la probabilidad de que esta sea de la urna 1?<br>

Consideremos los siguientes eventos:<br>
B: la bola extra√≠da es negra<br>
I: la urna elegida es la urna 1<br>

$P(I‚îÇB)=\frac{P(I‚à©B)}{P(B)}=\frac{P(B‚à©I)}{P(B)}=\frac{P(B|I)P(I)}{P(B‚à©I)+P(B‚à©I^C)}=\frac{\frac{1}{4}}{\frac{1}{4}+\frac{1}{2}\cdot\frac{2}{5}}= \frac{5}{9}$


```python
P_I = Fraction(1, 2)
P_Ic = Fraction(1, 2)

P_B_dado_I = Fraction(1, 2)
P_B_dado_Ic = Fraction(2, 5)

P_I_dado_B = (P_B_dado_I * P_I) / (P_B_dado_I * P_I + P_B_dado_Ic * P_Ic)

print(f"P(I|B) =", P_I_dado_B, "=", "{:.4f}".format(float(P_I_dado_B)))
```

### 12. LIDIA LA CONTADORA
Lidia realiza un examen de Contabilidad en el cual tiene que responder 3 preguntas que se encuentran escritas en tarjetas. Supongamos que hay 20 tarjetas<br> 
y Lidia elige aleatoriamente una de ellas. De las 20 tarjetas hay 8 tarjetas ‚Äúfavorables‚Äù (esto significa que Lidia conoce la respuesta de las 3 preguntas<br>
en cada una de estas tarjetas), todas las otras tarjetas contienen al menos una pregunta de para la cual Lidia no sabe la respuesta.<br>

La calificaci√≥n final de Lidia es aprobatoria si y solo si responde correctamente de las 3 preguntas de la tarjeta elegida.<br>

    (a) Cu√°l es la probabilidad de que Lidia apruebe si:
#### 1. ¬øEs la primera en sacar una tarjeta?

Sea el evento A: Lidia aprueba el examen

```python
total = 20
favorables = 8
no_favorables = total - favorables

P1 = Fraction(favorables, total)
print("P(1)=", P1 )
```

#### 2. ¬øEs la segunda en sacar una tarjeta?<br>

Antes de Lidia, alguien m√°s ya sac√≥ una tarjeta al azar, pero no sabemos cu√°l.<br>
Usamos la Ley de la Probabilidad Total:<br>

C1 = Caso 1: la primera tarjeta fue favorable<br>
C2 = Caso 2: la primera tarjeta fue no favorable<br>

$P(2)= P(A|C1)P(C1) + P(A|C2)P(C2)$<br>
„Ö§„Ö§$= \frac{7}{19} \cdot \frac{8}{20} + \frac{8}{19} \cdot \frac{12}{20}$<br>
„Ö§„Ö§$= \frac{56}{380} + \frac{96}{380}$<br> 
„Ö§„Ö§$= \frac{152}{380}$<br> 
„Ö§„Ö§$= \frac{2}{5}$<br>

```python
P2 = (Fraction(favorables, total) * Fraction(favorables - 1, total - 1) + Fraction(no_favorables, total) * Fraction(favorables, total - 1))
print("P(2)=", P2)
```

#### 3. ¬øEs la tercera en sacar una tarjeta?<br>
Ya se extrajeron dos tarjetas, por lo que consideramos lo siguiente:<br>
C1: las dos primeras tarjetas fueron favorables (FF)<br>
C2: la primera fue favorable y la segunda no favorable (FN)<br>
C3: la primera fue no favorable y la segunda favorable (NF)<br>
C4: las dos primeras fueron no favorables (NN)<br>

$P(3)= P(A|C1)P(C1) + P(A|C2)P(C2)+P(A|C3)P(C3)+P(A|C4)P(C4)$<br>
„Ö§„Ö§„Ö§$= \frac{6}{18}\cdot(\frac{8}{20})(\frac{7}{19}) + \frac{7}{18} \cdot (\frac{8}{20})(\frac{12}{19}) + \frac{7}{18} \cdot (\frac{12}{20})(\frac{8}{19}) + \frac{8}{18} \cdot (\frac{12}{20})(\frac{11}{19})$<br>
„Ö§„Ö§„Ö§$= \frac{14}{285} + \frac{28}{285}+ \frac{28}{285} + \frac{44}{285}$<br>
„Ö§„Ö§„Ö§$= \frac{2}{5}$

```python
P3 = (Fraction(favorables, total) *
    Fraction(favorables - 1, total - 1) *
    Fraction(favorables - 2, total - 2)
    +

    Fraction(favorables, total) *
    Fraction(no_favorables, total - 1) *
    Fraction(favorables - 1, total - 2)
    +

    Fraction(no_favorables, total) *
    Fraction(favorables, total - 1) *
    Fraction(favorables - 1, total - 2)
    +

    Fraction(no_favorables, total) *
    Fraction(no_favorables - 1, total - 1) *
    Fraction(favorables, total - 2))

print("P(3)=", P3)
```
